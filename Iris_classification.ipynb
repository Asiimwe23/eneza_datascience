## Iris Classification

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
iris_df['Species'] = iris['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=6)
knn_model = KNeighborsClassifier()
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
knn_model.fit(X_train, y_train)
```

```python
print("The accuracy on the test data:", 100*knn_model.score(X_test, y_test))
```

```python
print("The accuracy on the training data:", 100*knn_model.score(X_train, y_train))
```

When the perfomance on the training data is better than on the test data, the model is overfitting.
When the perfomance on the test data is better than on the training data, the model is generalising well (as long as both perfomances are pretty good).
When both the perfomances on both the training and test data is poor, the model is underfitting.
